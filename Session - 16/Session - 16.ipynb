# Task 1.
Given the following confusion matrix from a spam filter (Actual: Spam=70, Not Spam=130; Predicted: Spam=60, Not Spam=140; TP=50, TN=120, FP=10, FN=20), manually calculate the values of accuracy, precision, recall, and F1 score on paper or in a Jupyter notebook.

```python
tp = 50
tn = 120
fp = 10
fn = 20
```

```python
accuracy = (tp + tn) / (tp + tn + fp + fn)
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)
```

```python
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
```

# Task 4.
Given a confusion matrix for a food delivery fraud detection system (TP=40, TN=150, FP=20, FN=10), write Python code to compute the F1 score and explain in one line what the F1 score tells you about the model's performance.<br><br><em><strong>Hint:</strong> F1 = 2 * (precision * recall) / (precision + recall)</em>

```python
tp_fraud = 40
tn_fraud = 150
fp_fraud = 20
fn_fraud = 10
```

```python
precision_fraud = tp_fraud / (tp_fraud + fp_fraud)
recall_fraud = tp_fraud / (tp_fraud + fn_fraud)

f1_fraud = 2 * (precision_fraud * recall_fraud) / (precision_fraud + recall_fraud)
```

```python
print(f"Fraud Detection F1 Score: {f1_fraud:.4f}")
```

**Explanation:**
The F1 score is the harmonic mean of precision and recall, offering a single balanced metric that penalizes extreme values; an F1 score of ~0.72 means the fraud detection model is fairly reliable at catching fraud without making too many false accusations.
