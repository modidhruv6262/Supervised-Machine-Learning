# Session - 12 Classification Algorithms Part 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Task 1.
Load the Iris dataset using scikit-learn, train a DecisionTreeClassifier, and plot the resulting decision tree using plot_tree().

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(15, 10))
plot_tree(tree_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Full Depth Decision Tree - Iris Dataset")
plt.show()
```

# Task 2.
Change the splitting criterion of your DecisionTreeClassifier from 'gini' to 'entropy' and compare the accuracy scores for both on the same dataset.<br><br><em><strong>Hint:</strong> Use the 'criterion' parameter when initializing the classifier.</em>

```python
gini_preds = tree_clf.predict(X_test)
gini_acc = accuracy_score(y_test, gini_preds)
```

```python
# Entropy
entropy_clf = DecisionTreeClassifier(criterion='entropy', random_state=42)
entropy_clf.fit(X_train, y_train)
```

```python
entropy_preds = entropy_clf.predict(X_test)
entropy_acc = accuracy_score(y_test, entropy_preds)
```

```python
print(f"Accuracy using 'gini':    {gini_acc:.4f}")
print(f"Accuracy using 'entropy': {entropy_acc:.4f}")
```

**Comparison:**  
Both `gini` and `entropy` often produce very similar or identical accuracy on standard datasets like Iris. Gini measures the probability of misclassifying a random element, while Entropy measures information gain. Gini is the default because it is slightly faster to compute (no logarithms required).

# Task 3.
Limit the maximum depth of your decision tree to 2 and visualize the new tree structure. Observe and briefly describe how the tree's decisions have changed compared to the full-depth tree.

```python
pruned_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
pruned_clf.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 6))
plot_tree(pruned_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Pruned Decision Tree (max_depth=2)")
plt.show()
```

**Observation:**  
The full-depth tree created many complex splits to perfectly classify every single training sample (which can lead to overfitting). By limiting `max_depth=2`, the tree is forced to stop early. It only makes two levels of decisions. While it might misclassify a few edge cases (leaves are no longer completely "pure"), it creates a much simpler, more generalized rule set that is easier to interpret.
