## 2.
Given the following scenarios, classify each as supervised, unsupervised, or reinforcement learning: 1) Netflix recommending movies based on your watch history, 2) Spotify grouping similar songs into playlists, 3) A self-driving car learning to park by trial and error. Write your answers with a one-line explanation for each choice.

### 1. Netflix recommending movies based on your watch history
**Classification:** Supervised Learning   
**Explanation**: The model uses your labeled historical data (movies you have watched, rated, or liked) to predict what you will want to watch next.

### 2. Spotify grouping similar songs into playlists
**Classification**: Unsupervised Learning   
**Explanation**: The algorithm clusters tracks together based on natural, inherent acoustic patterns and user habits without using any pre-defined labels or categories.

### 3. A self-driving car learning to park by trial and error
**Classification**: Reinforcement Learning    
**Explanation**: The autonomous system learns optimal driving maneuvers through direct interaction with its environment by earning rewards for success and penalties for mistakes.

## 4.
Imagine you are building a spam filter for WhatsApp messages. Write a short paragraph explaining how traditional programming would approach this problem versus how machine learning would approach it.<br><br><em><strong>Hint:</strong> Focus on rules vs learning from data.</em>

**Traditional programming** requires a human to write strict, manual rules (like blocking the word "lottery"). If spammers change the spelling to "l0ttery," the system breaks until a programmer manually updates the code.

**Machine learning** automatically learns by looking at data. You give it thousands of examples labeled as "spam" or "not spam," and the algorithm figures out the patterns on its own. It dynamically adapts to new tricks without needing any manual code changes.
