# Task 4.
Take a Zomato restaurant dataset and use scikit-learn's RFE (Recursive Feature Elimination) with a DecisionTreeRegressor to select the 3 most important features for predicting 'average cost for two'. List the selected features and the RFE ranking.

```python
df = pd.read_csv(r"C:\Users\DHRUV MODI\Documents\TOPS\Data set\zomato restaurant.csv")
df.head()
```

```python
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0})
df['table booking'] = df['table booking'].map({'Yes': 1, 'No': 0})
```

```python
df.dropna(inplace=True)
```

```python
X = df[['rate (out of 5)', 'num of ratings', 'online_order', 'table booking']]
y = df['avg cost (two people)']
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()
```

```python
rfe = RFE(estimator=model, n_features_to_select=3)
rfe.fit(X, y)
```

```python
rfe.ranking_
```

# Task 5.
Given a BookMyShow movie ratings dataset, detect outliers in the 'user ratings' column using the IQR method. Remove the outliers and display how many rows were dropped.<br><br><em><strong>Hint:</strong> Calculate Q1, Q3, and IQR; filter out rows outside [Q1 - 1.5*IQR, Q3 + 1.5*IQR].</em>

```python
df=pd.read_csv(r"C:\Users\DHRUV MODI\Documents\TOPS\Data set\movie dataset rating\Movie-Dataset-Latest.csv")
df.head()
```

```python
q1 = df['vote_average'].quantile(0.25)
q3 = df['vote_average'].quantile(0.75)
iqr = q3 - q1
```

```python
q1,q3,iqr
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
lower_bound,upper_bound
```

```python
df_filtered = df[(df['vote_average'] >= lower_bound) & (df['vote_average'] <= upper_bound)]
df_filtered
```

```python
rows_dropped = len(df) - len(df_filtered)
rows_dropped
```
