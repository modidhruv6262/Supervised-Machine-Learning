# Task 1.
Create a Python function sigmoid(x) that takes a numeric input and returns the sigmoid value using the formula 1 / (1 + exp(-x)). Test your function with x = -2, 0, and 3.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
print(f"sigmoid(-2) = {sigmoid(-2)}")
print(f"sigmoid(0)  = {sigmoid(0)}")
print(f"sigmoid(3)  = {sigmoid(3)}")
```

# Task 3.
Plot the sigmoid function for x values from -10 to 10 using matplotlib, and visually indicate the threshold at 0.5 on the plot.<br><br><em><strong>Hint:</strong> Use plt.axhline() to draw the threshold line.</em>

```python
x_values = np.linspace(-10, 10,300)
y_values = sigmoid(x_values)
```

```python
plt.plot(x_values, y_values, color='blue', label='Sigmoid')
plt.axhline(y=0.5, color='red', linestyle='--', label='Threshold = 0.5')
plt.axvline(x=0, color='gray', linestyle=':', linewidth=0.8)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.title("Sigmoid Function")
plt.legend()
plt.show()
```

# Task 4.
Given a probability output from your logistic regression model, write a function that classifies the result as 'Viral' or 'Not Viral' based on a threshold value that you can set (default 0.5). Test your function with probabilities 0.2, 0.6, and 0.8 and thresholds 0.5 and 0.7.

```python
def classify(probability, threshold=0.5):
    if probability >= threshold:
        return 'Viral'
    else:
        return 'Not Viral'
```

```python
probabilities = [0.2, 0.6, 0.8]
thresholds = [0.5, 0.7]

for threshold in thresholds:
    print(f"\nThreshold = {threshold}")
    for prob in probabilities:
        result = classify(prob, threshold)
        print(f"  probability={prob} → {result}")
```

**Key Observation:**   
At a **threshold of 0.5**, probabilities 0.6 and 0.8 are classified as `Viral` while 0.2 is `Not Viral`. When the threshold is raised to **0.7**, only a probability of 0.8 qualifies as `Viral` — the model becomes stricter. In real-world Instagram analytics, a higher threshold means only posts the model is very confident about are labelled viral, reducing false positives but potentially missing some truly viral posts.
