# Task 1.
Load the Iris dataset using scikit-learn and use cross_val_score() with a KNeighborsClassifier to compute the average accuracy using 5-fold cross-validation.

```python
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

```python
iris = load_iris()
X_iris, y_iris = iris.data, iris.target

knn_model = KNeighborsClassifier(n_neighbors=3)
```

```python
knn_scores = cross_val_score(knn_model, X_iris, y_iris, cv=5)
print(f"Average 5-fold Accuracy: {knn_scores.mean():.4f}")
```

# Task 4.
Experiment with different values of 'cv' (3, 5, 10) in cross_val_score() for a RandomForestClassifier on the Wine dataset and report which value gives the most stable (least variable) accuracy scores.<br><br><em><strong>Hint:</strong> Look at the standard deviation of the scores for each cv value.</em>

```python
from sklearn.datasets import load_wine
from sklearn.ensemble import RandomForestClassifier
```

```python
wine = load_wine()
X_wine, y_wine = wine.data, wine.target

rf_wine = RandomForestClassifier(random_state=42)
```

```python
cv_options = [3, 5, 10]

for cv_val in cv_options:
    scores = cross_val_score(rf_wine, X_wine, y_wine, cv=cv_val)
    print(f"CV={cv_val:2} | Mean: {scores.mean():.4f} | Std Dev: {scores.std():.4f}")
```

**Explanation:**
Typically, a highly stable score means a low Standard Deviation (Std Dev). While CV=10 often gives a highly accurate mean, it can have higher variance because the test sets in each fold are very small. CV=5 usually provides the best "goldilocks" balance of stability and accuracy for small datasets like Wine!

Use ChatGPT or Copilot to help you write a function that takes any classifier and dataset, runs k-fold cross-validation using cross_val_score(), and returns the mean and standard deviation of the scores. Paste the code you generated and briefly describe how the AI tool improved your solution.

```python
# AI Generated Code block for a robust CV evaluator
def evaluate_model_cv(model, X, y, cv_folds=5):
    """
    Evaluates a model using cross-validation and returns mean & std dev.
    """
    scores = cross_val_score(model, X, y, cv=cv_folds)
    mean_acc = scores.mean()
    std_dev = scores.std()
    
    return mean_acc, std_dev
```

```python
mean, std = evaluate_model_cv(rf_wine, X_wine, y_wine, cv_folds=5)
print(f"Function Result -> Mean: {mean:.4f}, Std: {std:.4f}")
```
